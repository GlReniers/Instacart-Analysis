## Task 4.8

```python
# Check statistics of product prices for each loyalty category (number limits placed by data scientists)

ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean','min','max']})
```

```python
ords_prods_merge.shape
```

```python
ords_prods_merge.head(20)
```

```python
# 1. Create new column, mean price by user
ords_prods_merge['mean_order']= ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# 2. Assign spending flag based on average price

ords_prods_merge.loc[ords_prods_merge['mean_order']<10, 'spending_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['mean_order']>=10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

```python
ords_prods_merge[['user_id', 'mean_order', 'spending_flag']].head(100)
```

```python
#Create a frequency flag based on median days since previous order was made.
# 1.  Create new column for median days since previous order

ords_prods_merge['median_days_since_previous_order'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.median)
```

```python
ords_prods_merge.shape
```

```python
ords_prods_merge.head()
```

```python
# 2. Assign frequency flags based on median days since previous order
ords_prods_merge.loc[ords_prods_merge['median_days_since_previous_order'] > 20, 'Frequency_flag'] = 'New customer'
ords_prods_merge.loc[(ords_prods_merge['median_days_since_previous_order'] > 10) & (ords_prods_merge['median_days_since_previous_order'] <= 20), 'Frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_days_since_previous_order'] <= 10, 'Frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.shape
```

```python
ords_prods_merge['Frequency_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'median_days_since_previous_order', 'Frequency_flag']].head(100)
```

```python
# Renaming mean_order column to mean_price to make it easier to understand
ords_prods_merge.rename(columns = {'mean_order':'mean_price'}, inplace = True)
```
